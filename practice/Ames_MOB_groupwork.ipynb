```python
house_mob = train[['Lot Area', 'Year Built', 'Mo Sold', 'Yr Sold', 'Bath', 'SalePrice','Gr Liv Area','Total Bsmt SF', 'Bedroom AbvGr']]
```

```python
from matplotlib.patches import Rectangle


plt.figure(figsize = (16, 10))

mask = np.zeros_like(house_mob.corr())
mask[np.triu_indices_from(mask)] = True

ax = sns.heatmap(house_mob.corr(), mask=mask, vmin=0.2, annot=True, cmap="YlGnBu");
ax.add_patch(Rectangle((5, 6), 1, 1, fill=False, edgecolor='red', lw=8));
ax.add_patch(Rectangle((1, 5), 1, 1, fill=False, edgecolor='orange', lw=8));
```
